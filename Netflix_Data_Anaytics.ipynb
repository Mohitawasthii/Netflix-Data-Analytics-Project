#- Handle Missing Values:

**Insight:** Missing values were handled to ensure the dataset is complete and suitable for analysis.
Categorical columns such as director, cast, and country were filled with placeholder values like 'Unknown' or 'Not Available' to retain important records.
Missing values in columns such as rating, duration, and date_added were replaced using the most frequent value (mode) to maintain data consistency.

**Handling missing values ensures that the Netflix dataset remains comprehensive and reliable, leading to more accurate analysis and meaningful business insights.**

```python
data['director'] = data['director'].fillna('Unknown')
data['cast'] = data['cast'].fillna('Not Available')
data['country'] = data['country'].fillna('Unknown')

data['date_added'] = data['date_added'].fillna(
    data['date_added'].mode()[0])

data['rating'] = data['rating'].fillna(
    data['rating'].mode()[0])

data['duration'] = data['duration'].fillna(
    data['duration'].mode()[0])
```

#Outlier Detection:

**Insight:** Outlier detection was performed using the Interquartile Range (IQR) method on the release_year variable.
Very old titles may appear as statistical outliers; however, they represent genuine historical content available on Netflix.
Since these records are valid observations rather than data entry errors, outlier capping is preferred over removal to preserve valuable information.

```python
plt.figure(figsize=(10,4))

sns.boxplot(x=data['release_year'])

plt.title('Release Year Distribution')

plt.show()
```

#-Handling Outliers:
The Netflix dataset contains very few numerical columns. The primary numerical feature is release_year.

#Calculate IQR:

```python
Q1 = data['release_year'].quantile(0.25)
Q3 = data['release_year'].quantile(0.75)

IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

print("Lower Bound:", lower_bound)
print("Upper Bound:", upper_bound)
```

#Remove Outliers:

```python
data = data[
    (data['release_year'] >= lower_bound) &
    (data['release_year'] <= upper_bound)
]
```

#Cap Outliers:
Instead of removing records, cap extreme values.


```python
data['release_year'] = np.where(
    data['release_year'] > upper_bound,
    upper_bound,
    data['release_year']
)

data['release_year'] = np.where(
    data['release_year'] < lower_bound,
    lower_bound,
    data['release_year']
)
```

#After Handling Outliers:


```python
plt.figure(figsize=(10,5))

sns.boxplot(x=data['release_year'])

plt.title('Boxplot After Handling Outliers')

plt.show()
```

# -Duration Analysis of Movies and TV Shows:

**Insight:** Most Netflix movies have durations ranging between 80 and 120 minutes, indicating a preference for standard feature-length films.
The average movie duration is approximately 100 minutes, which aligns with industry standards.
Very short and extremely long movies are comparatively rare in the Netflix catalog.
The majority of TV Shows on Netflix consist of only one season, suggesting a strong presence of limited series and mini-series.

```python
movies = data[data['type'] == 'Movie'].copy()

movies['minutes'] = movies['duration'].str.replace(' min', '')

movies['minutes'] = pd.to_numeric(movies['minutes'], errors='coerce')

movies.dropna(subset=['minutes'], inplace=True)

movies['minutes'] = movies['minutes'].astype(int)
```

#-Boxplot:

**Insight:** Outliers in movie duration were identified using the Interquartile Range (IQR) method.
A few movies exhibit exceptionally long durations and appear as outliers in the distribution.


```python
plt.figure(figsize=(10,5))

sns.boxplot(x=movies['minutes'])

plt.title('Movie Duration Boxplot')

plt.show()
```

#Calculate IQR

```python
Q1 = movies['minutes'].quantile(0.25)
Q3 = movies['minutes'].quantile(0.75)

IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

print("Lower Bound:", lower_bound)
print("Upper Bound:", upper_bound)
```

#Detect Outliers:

```python
outliers = movies[
    (movies['minutes'] < lower_bound) |
    (movies['minutes'] > upper_bound)
]

print("Number of Outliers:", outliers.shape[0])
```

#Remove Outliers:

```python
movies = movies[
    (movies['minutes'] >= lower_bound) &
    (movies['minutes'] <= upper_bound)
]
```

```python
movies = data[data['type'] == 'Movie'].copy()

movies['minutes'] = movies['duration'].str.replace(
    ' min', ''
)

movies['minutes'] = pd.to_numeric(movies['minutes'], errors='coerce')
movies.dropna(subset=['minutes'], inplace=True)

movies['minutes'] = movies['minutes'].astype(int)
```
